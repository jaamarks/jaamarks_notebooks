This document logs the steps taken to perform pre-imputation procedures on the merged datasets of UHS1, UHS2, and UHS3 both EA and AA. The starting point for this analysis is after quality control of observed genotypes. The quality controlled genotypes are oriented on the GRCh37 plus strand. Based on findings from [Johnson et al.](https://link.springer.com/article/10.1007/s00439-013-1266-7), the intersection set of variants for will be used for imputation.

## Data processing
### GRCh37 strand and allele discordance check

````r
# EC2 command line #
cd /home/ec2-user/jmarks/hiv/uhs123_impute
base_dir=/home/ec2-user/jmarks/hiv/uhs123_impute

mkdir 1000g

for study in {UHS1,UHS2,UHS3_v1-2,UHS3_v1-3}; do
    mkdir -p ${base_dir}/${study}/genotype/strand_check
    for ancestry in {ea,aa};do
        /shared/bioinformatics/software/third_party/plink-1.90-beta-4.10-x86_64/plink \
            --noweb \
            --memory 2048 \
            --bed ${base_dir}/${study}/genotype/observed/${ancestry}/*bed \
            --bim ${base_dir}/${study}/genotype/observed/${ancestry}/*bim \
            --fam ${base_dir}/${study}/genotype/observed/${ancestry}/*fam \
            --freq \
            --out ${base_dir}/${study}/genotype/strand_check/${ancestry}_chrX
    done
done


# Get list of variants from all studies
for ancestry in {ea,aa};do
    cat ${base_dir}/{UHS1,UHS2,UHS3_v1-2,UHS3_v1-3}/genotype/observed/${ancestry}/*bim | \
            perl -lane 'if (($F[0]+0) <= 23) { print $F[1]; }' | \
            sort -u > ${base_dir}/${ancestry}_chrX_sorted_variants.txt
done

 wc -l *_chrX_sorted_variants.txt
"""
55030 aa_chrX_sorted_variants.txt
54454 ea_chrX_sorted_variants.txt
"""


# Calculate chrX MAFs for 1000G EUR and AFR
chr=23
for ancestry in {ea,aa};do
    if [ $ancestry == 'aa' ]
    then
        pop="AFR"
        echo $pop
    else
        pop="EUR"
        echo $pop
    fi

    /shared/bioinformatics/software/scripts/qsub_job.sh \
        --job_name ${pop}_${chr} \
        --script_prefix ${base_dir}/${pop}_chr${chr}.maf \
        --mem 7 \
        --priority 0 \
        --program perl /shared/bioinformatics/software/perl/stats/calculate_maf_from_impute2_hap_file.pl \
            --hap /shared/data/ref_panels/1000G/2014.10/1000GP_Phase3_chrX_NONPAR.hap.gz\
            --legend /shared/data/ref_panels/1000G/2014.10/1000GP_Phase3_chrX_NONPAR.legend.gz \
            --sample /shared/data/ref_panels/1000G/2014.10/1000GP_Phase3.sample \
            --chr ${chr} \
            --extract ${base_dir}/${ancestry}_chrX_sorted_variants.txt \
            --out ${base_dir}/1000g/${pop}_chr${chr}.maf \
            --keep_groups ${pop}
done

"""
wc -l 1000g/*
  51167 1000g/AFR_chr23.maf
  50855 1000g/EUR_chr23.maf
"""


# # Merge per chr MAFs for EUR
# for pop in {EUR,AFR]};do
#     head -n 1 ${base_dir}/1000g/${pop}_chr1.maf > ${base_dir}/1000g/${pop}_chr_all.maf
#     tail -q -n +2 ${base_dir}/1000g/${pop}_chr{1..23}.maf \
#         >> ${base_dir}/1000g/${pop}_chr_all.maf
# done

# Run discordance checks for EA and AA groups
for study in {UHS1,UHS2,UHS3_v1-2,UHS3_v1-3}; do
    for ancestry in {ea,aa};do
        if [ $ancestry == "ea" ]
        then
            pop="EUR"
        else
            pop="AFR"
        fi
    /shared/bioinformatics/software/scripts/qsub_job.sh \
        --job_name ${ancestry}_${study}_crosscheck \
        --script_prefix ${base_dir}/${study}/genotype/strand_check/${ancestry}_allele_discordance_check \
        --mem 6 \
        --priority 0 \
        --program "Rscript /shared/bioinformatics/software/R/check_study_data_against_1000G.R
            --study_bim_file ${base_dir}/${study}/genotype/observed/${ancestry}/*.bim
            --study_frq_file ${base_dir}/${study}/genotype/strand_check/${ancestry}_chrX.frq
            --ref_maf_file ${base_dir}/1000g/${pop}_chr23.maf
            --out_prefix ${base_dir}/${study}/genotype/strand_check/${ancestry}_allele_discordance"
done
````

### Resolving allele discordances
The allele discordances will be resolved by
* Flipping allele discordances that are fixed by flipping
* Removing SNPs with discordant names
* Removing SNPs with discordant positions
* Removing allele discordances that are not resolved by flipping
* Removing alleles with large deviations from the reference population allele frequencies

Given that the allele discordance check was done using a union set of SNPs across all studies within an ancestry group, some of the SNPs logged as discordant for a given study may not actually be in the study. Fortunately, if they are not in a given study they will not interfere with the filtering procedures.

````r
# EC2 command line #
cd /home/ec2-user/jmarks/hiv/uhs123_impute
base_dir=/home/ec2-user/jmarks/hiv/uhs123_impute

for study in {UHS1,UHS2,UHS3_v1-2,UHS3_v1-3}; do
    for ancestry in {ea,aa};do
        echo -e "\n===============\nProcessing ${study}\n"
        # Create remove list
        echo "Making remove list"
        cat <(cut -f2,2 ${base_dir}/${study}/genotype/strand_check/${ancestry}_allele_discordance.discordant_alleles_not_fixed_by_strand_flip | tail -n +2) \
            <(cut -f2,2 ${base_dir}/${study}/genotype/strand_check/${ancestry}_allele_discordance.at_cg_snps_freq_diff_gt_0.2 | tail -n +2) \
            <(cut -f2,2 ${base_dir}/${study}/genotype/strand_check/${ancestry}_allele_discordance.discordant_names | tail -n +2) \
            <(cut -f2,2 ${base_dir}/${study}/genotype/strand_check/${ancestry}_allele_discordance.discordant_positions | tail -n +2) \
            <(cut -f2,2 ${base_dir}/${study}/genotype/strand_check/${ancestry}_allele_discordance.discordant_alleles_polymorphic_in_study_not_fixed_by_strand_flip | tail -n +2) | \
            sort -u > ${base_dir}/${study}/genotype/${ancestry}_snps.remove

        # Create flip list
        echo "Making flip list"
        comm -23 <(cut -f2,2 ${base_dir}/${study}/genotype/strand_check/${ancestry}_allele_discordance.discordant_alleles | tail -n +2 | sort -u) \
            <(cut -f2,2 ${base_dir}/${study}/genotype/strand_check/${ancestry}_allele_discordance.discordant_alleles_not_fixed_by_strand_flip | tail -n +2 | sort -u) \
            > ${base_dir}/${study}/genotype/${ancestry}_snps.flip

        # Apply filters
        /shared/bioinformatics/software/third_party/plink-1.90-beta-4.10-x86_64/plink \
            --noweb \
            --memory 2048 \
            --bed ${base_dir}/${study}/genotype/observed/${ancestry}/*bed \
            --bim ${base_dir}/${study}/genotype/observed/${ancestry}/*bim \
            --fam ${base_dir}/${study}/genotype/observed/${ancestry}/*fam \
            --exclude ${base_dir}/${study}/genotype/${ancestry}_snps.remove \
            --flip ${base_dir}/${study}/genotype/${ancestry}_snps.flip \
            --make-bed \
            --out ${base_dir}/${study}/genotype/observed/$ancestry/${ancestry}_filtered
    done
done

"""
 wc -l */genotype/ob*/*/*bim
 wc -l */genotype/ob*/{aa,ea}/*bim
   17244 UHS1/genotype/observed/aa/aa_filtered.bim
   17396 UHS1/genotype/observed/aa/uhs1.aa.chr23.bim
   39367 UHS2/genotype/observed/aa/aa_filtered.bim
   39502 UHS2/genotype/observed/aa/uhs2.aa.chr23.bim
   40158 UHS3_v1-2/genotype/observed/aa/aa_filtered.bim
   41884 UHS3_v1-2/genotype/observed/aa/uhs3.aa.V1-2.chr23.bim
   43426 UHS3_v1-3/genotype/observed/aa/aa_filtered.bim
   44706 UHS3_v1-3/genotype/observed/aa/uhs3.aa.V1-3.chr23.bim
   17255 UHS1/genotype/observed/ea/ea_filtered.bim
   17408 UHS1/genotype/observed/ea/uhs1.ea.chr23.bim
   45738 UHS2/genotype/observed/ea/ea_filtered.bim
   45850 UHS2/genotype/observed/ea/uhs2.ea.chr23.bim
   23834 UHS3_v1-2/genotype/observed/ea/ea_filtered.bim
   24202 UHS3_v1-2/genotype/observed/ea/uhs3.ea.V1-2.chr23.bim
   34753 UHS3_v1-3/genotype/observed/ea/ea_filtered.bim
   35440 UHS3_v1-3/genotype/observed/ea/uhs3.ea.V1-3.chr23.bim

"""
````

### Remove monomorphic variants
Monomorphic variants prevent MIS from accepting the genotype data. In this case, an arbitrarily small MAF is set that is smaller than the lower bound for these data.

````r
# EC2 command line #
cd /home/ec2-user/jmarks/hiv/uhs123_impute
base_dir=/home/ec2-user/jmarks/hiv/uhs123_impute

# Apply filters
for study in {UHS1,UHS2,UHS3_v1-2,UHS3_v1-3}; do
    for ancestry in {ea,aa};do
        /shared/bioinformatics/software/third_party/plink-1.90-beta-4.10-x86_64/plink \
            --noweb \
            --memory 2048 \
            --bfile ${base_dir}/${study}/genotype/observed/$ancestry/${ancestry}_filtered \
            --maf 0.000001 \
            --make-bed \
            --out ${base_dir}/${study}/genotype/observed/$ancestry/${ancestry}_filtered_mono
    done
done

"""
wc -l ${base_dir}/*/genotype/observed/{ea,aa}/*bim
   17255 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS1/genotype/observed/ea/ea_filtered.bim
   15532 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS1/genotype/observed/ea/ea_filtered_mono.bim
   17408 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS1/genotype/observed/ea/uhs1.ea.chr23.bim
   45738 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS2/genotype/observed/ea/ea_filtered.bim
   34665 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS2/genotype/observed/ea/ea_filtered_mono.bim
   45850 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS2/genotype/observed/ea/uhs2.ea.chr23.bim
   23834 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS3_v1-2/genotype/observed/ea/ea_filtered.bim
   19704 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS3_v1-2/genotype/observed/ea/ea_filtered_mono.bim
   24202 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS3_v1-2/genotype/observed/ea/uhs3.ea.V1-2.chr23.bim
   34753 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS3_v1-3/genotype/observed/ea/ea_filtered.bim
   26381 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS3_v1-3/genotype/observed/ea/ea_filtered_mono.bim
   35440 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS3_v1-3/genotype/observed/ea/uhs3.ea.V1-3.chr23.bim
   17244 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS1/genotype/observed/aa/aa_filtered.bim
   16395 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS1/genotype/observed/aa/aa_filtered_mono.bim
   17396 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS1/genotype/observed/aa/uhs1.aa.chr23.bim
   39367 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS2/genotype/observed/aa/aa_filtered.bim
   35951 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS2/genotype/observed/aa/aa_filtered_mono.bim
   39502 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS2/genotype/observed/aa/uhs2.aa.chr23.bim
   40158 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS3_v1-2/genotype/observed/aa/aa_filtered.bim
   37063 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS3_v1-2/genotype/observed/aa/aa_filtered_mono.bim
   41884 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS3_v1-2/genotype/observed/aa/uhs3.aa.V1-2.chr23.bim
   43426 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS3_v1-3/genotype/observed/aa/aa_filtered.bim
   39760 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS3_v1-3/genotype/observed/aa/aa_filtered_mono.bim
   44706 /home/ec2-user/jmarks/hiv/uhs123_impute/UHS3_v1-3/genotype/observed/aa/uhs3.aa.V1-3.chr23.bim
"""
````

### Merge test
