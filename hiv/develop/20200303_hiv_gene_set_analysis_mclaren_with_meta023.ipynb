## Add Gene Symbols
Also convert to tab-separated files.

```r
cd /shared/jmarks/hiv/gene-set-analysis/final
for file in *_gene_analysis.genes.out; do
    inFile=$file
    outFile=$file.tsv
        head -n 1 $inFile |
            perl -lne '
                s/\s+/\t/g;
                /^(\S+)\s+(.+)/;
                print join("\t",$1,"SYMBOL",$2);
                ' > $outFile
        tail -n +2 $inFile |
            perl -lne '
                BEGIN {
                    %symbols = ();
                    open(XREF, "/shared/bioinformatics/software/third_party/magma_v1.07b/gene_loc/NCBI37.3/NCBI37.3.gene.loc");
                    while (<XREF>) {
                    /(\S+)(?:\s+\S+){4}\s+(\S+)/;
                        $symbols{$1} = $2;
                    }
                    close XREF;
                }
                s/\s+/\t/g;
                /^(\S+)\s+(.+)/;
                if (exists($symbols{$1})) {
                    print join("\t",$1,$symbols{$1},$2);
                } else {
                    print join("\t",$1,"",$2);
                }' >> $outFile
done
```
