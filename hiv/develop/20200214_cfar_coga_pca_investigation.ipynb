# PC comparison between CFAR & COGA
Looking at the PCs that went into the GWAS model—PC1, PC3, and PC7.

```python
%matplotlib notebook
import matplotlib.pyplot as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
mytit = "CFAR+COGA PCs that went into the GWAS model"
plt.title(mytit, pad=15)
ax.set_xlabel("PC1 (36%)", labelpad=10) # 25.098/69.34
ax.set_ylabel("PC3 (12%)", labelpad=10) # 5.347/69.34
ax.set_zlabel("PC7 (6%)", labelpad=10)  # 4.266/69.34

import os

os.chdir("/Users/jmarks/OneDrive - Research Triangle Institute/Projects/hiv/cfar_coga/troubleshoot/pca_comparison/")
pca_df = pd.read_csv("cfar_coga_pca_df.txt", sep=" ", index_col=0, header=0)

# cfar
cfar = pca_df.loc["Case",:]
pc1 = cfar["EV1"]
pc3 = cfar["EV3"]
pc7 = cfar["EV7"]
cfar_scatter = ax.scatter(pc1, pc3, pc7, color="green")


# coga
coga = pca_df.loc["Control",:]
pc1 = coga["EV1"]
pc3 = coga["EV3"]
pc7 = coga["EV7"]
coga_scatter = ax.scatter(pc1, pc3, pc7, color="blue")



leg = ax.legend((cfar_scatter, coga_scatter),
          ("cfar", "coga"), loc="upper left")
bb = leg.get_bbox_to_anchor().inverse_transformed(ax.transAxes)

# Change to location of the legend. 
xOffset = -0.10
yOffset = -0.05
bb.x0 += xOffset
bb.y0 += yOffset
bb.y1 += yOffset
#bb.x1 += xOffset
leg.set_bbox_to_anchor(bb, transform = ax.transAxes)
#for i, txt in enumerate(pca_df.index.values):
#    if txt == "Case":
#        mycol = "green"
#    else:
#        mycol = "blue"
#    ax.scatter(pc1.iloc(i),pc3.iloc(i),pc7.iloc(i), color=mycol)
#    ax.draw()
#    #ax.text(pc1[i], pc3[i], pc7[i], txt, fontsize=8)
    
#plt.show()
```
