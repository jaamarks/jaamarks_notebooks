```r
alt_allele <- read.delim(file = "/Users/jmarks/projects/nicotine/lookup/gscan/tables-6-9-variants-and-maf.txt",
                head = T)
class(alt_allele)
```

```r
x <- c(0.3, 0.6, 0.55)

alt_to_maf <- function(x) {
  # function that converts temperature in degrees Fahrenheit to kelvin
  # input: fahr: numeric value representing temp in degrees farh
  # output: kelvin: numeric converted temp in kelvin
    if (x > 0.5){
      maf <- 1 - x
    }
    else {
        maf <- x
    }
  return(maf)
}
#maf <- sapply(X = x, FUN = alt_to_maf)
maf <- sapply(X = alt_allele$Alternate.Allele.Frequnecy, FUN = alt_to_maf)
```

```r
maf <- as.data.frame(maf)
class(maf)
```

```r
mybreaks <- seq(from = 0, to = 1, by = 0.025)
hist(maf$maf, breaks = 16, col = "lightblue", axes = T,
    main = "Alternate allele frequency distribution\nof the genome-wide significant GSCAN SNPs",
    labels = T)
```
