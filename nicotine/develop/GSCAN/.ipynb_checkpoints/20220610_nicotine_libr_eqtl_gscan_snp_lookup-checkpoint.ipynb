````r
### Python3 ###
"""
*SNP lookup*

Search for the GSCAN SNPs within the maf_filtered_qtls LIBD results.
Note that I changed the name of the following GSCAN SNPs because the LIBD had old rsIDs.
rs12611472 --> rs147137183
rs28813180 --> rs138279803
"""
import gzip

################################################################################
################################################################################

gscan_table = "table1_ai.txt".format()
base_dir = "/shared/jmarks/eQTL_analysis/NAc_data/matrix_eqtl"  # no forward slash at the end
snp_list = "{}/gscan/{}".format(base_dir, gscan_table)

eqtl_basedir = "/shared/lzhou/eQTL_analysis/NAc_data/matrix_eqtl/output/baseline/adjusted_pvalues"

## results files reporting the gscan snps found and not found 
out_file = "{}/results/20220610/baseline/2df_mqtl_mapping_gscan_{}_lookup.txt".format(base_dir, gscan_table)
not_found = "{}/results/20220610/baseline/2df_mqtl_mapping_gscan_{}_lookup_snps_not_found.txt".format(base_dir, gscan_table)


head_line = "false" # write the header to outfile, then change to true so it won't print it again
with open(snp_list) as gscanF, open(out_file, "w") as outF:
    line = next(gscanF) # skip header
    line = gscanF.readline()
    # Chr     Pos     rsID
    # 2       145638766       rs72853300

    ## create a dictionary containing the gscan snps (rsIDs)
    gscan_dict =  {}
    while line:
        snp = line.split()[2]
        gscan_dict[snp] = 0 # initiate the number of times it has been found 
        line = gscanF.readline()

    print(gscan_dict)


    for chrom in range(2,9):
        which_chrom = "chr{}".format(chrom)
        print(which_chrom)
        mqtl_results = "{}/filtered_MEGA_MAF_0.05_two_stage_adjusted_pvalue_LIBD_NAc_MEGA_baseline_chr{}_no_smoking_status_eqtl_mapping_cis_qtl_table.txt".format(eqtl_basedir, chrom)

        with open(mqtl_results) as mqtlF:
            mqtl_head = mqtlF.readline()
            print(mqtl_head)
            #SNP     gene    beta    t-stat  p-value ...
            #rs707920        ENSG00000204428.12      1.05777341354105        21.7425255850819        2.54744984915941e-53...
            if head_line == "false":
                outF.write(mqtl_head)
                head_line = "true"

            mqtl_line = mqtlF.readline()
            while mqtl_line: # look at all of the variety
                variant_id = mqtl_line.split()[0] # SNP
                rsid = variant_id.split(":")[0]

                if rsid in gscan_dict:
                    gscan_dict[rsid] += 1 # mark the variant as found   
                    outF.write(mqtl_line)
                mqtl_line = mqtlF.readline()


with open(not_found, 'w') as notF:
    for item in gscan_dict.items():
        count = item[1]
        if count == 0:
            notF.write(item[0] + "\n")
````

# Random SNP selection
We will pull ~3050 random SNPs from the eQTL results and repeat the SNP lookup. We want to determine if the GSCAN lookup results reflected true biology. Performing a random SNP lookup will help us to that end by helping us answer the question "What is the probability that these two sets of samples were drawn from the same (but unknown) probability distribution?". We will perform a Kolmogorov–Smirnov test for this. 


We will use SNPsnap to select the random SNPs.

**SNPsnap**
```
SNPsnap efficiently identifies sets of randomly drawn SNPs that are matched to a set of query SNPs based on

    Minor allele frequency
    Number of SNPs in linkage disequilibrium (LD buddies)
    Distance to nearest gene
    Gene density
```

Select 3050 random SNPs for a SNP enrichment test. Need to make sure the random SNPs:
* are in linkage equilibrium with each other
* are in linkage equilibrium with GSCAN.

**The approach we will take is as follows:**
1. Create a list of unique GSCAN SNPs that were found in the eQTL results.
1. Generated a host of random SNPs using the SNPsnap program.
1. Pull a set of ~4000 of those SNPs based on LD and their distribution across the genome to match the distribution of the 305 GSCAN SNPs.
1. Prune those ~4000 random SNPs down based on LD
1. Prune the 305 GSCAN SNPs down based on LD.
1. Combined the random LE SNPs and the GSCAN LE SNPs, then test for LD between these combined SNPs.<br>
    a. indicate that we would like to retain the GSCAN SNPs in the final pruned set.
1. Remove the GSCAN SNPs from the resulting ld-pruned random+GSCAN set. Your left with just ~3,000 random SNPs.
1. Randomly select 3,050 of these SNPs.

## Select subset of random SNPs
In the end we want 3050 random SNPs, but some of the random SNPs we pull will be in LD. So we will pull more than 3050 so that when we perform LD we will still have >3,050 SNPs to choose from.

````r
cd ../
````

````r
### python3
"""
Select subset of random SNPs from the large set of 304×10,000 random SNPs.
1 SNP from the 305 was not found in the SNPsnap program. 

Select 15 random SNPs for every input SNP for a total of 15×304 = 4,560 random SNPs
* 4494 random SNPs were actually found that were in the LIBD genotype data

Check to make sure the random SNP is actually in the LIBD genotype data. Use the 
rsID list we created earlier.

Make sure you don't select a random SNP more than once.
"""

infile = "SNPsnap_305_gscan_input_v01/matched_snps.txt"
outfile = "matched_4494_random_snps.txt"
libd_list = "aa_libd_variant_ids_and_rsid.txt"

with open(infile) as inF, open(libd_list) as libdF, open(outfile, 'w') as outF:
    line = libdF.readline()
    print(line) # rs140435168:256586:T:G  1:256586

    libd_dict = dict()
    # make sure the random SNPs we pick are actually in the LIBD genotype data
    while line:
        sl = line.split()
        libd_dict[sl[1]] = sl[0] # chr:pos=key  rsid=value 
        line = libdF.readline()


    head = inF.readline()
    outhead = "gscan_snp\trandom_id\trandom_rsid\n"
    outF.write(outhead)
    line = inF.readline()

    dup_set = set() # make sure not to pick a random SNP >1 time
    while line: # loop thru all 304 GSCAN SNPs
        sl = line.split()
        gscan_snp = sl[0] # chr:pos
        gscan_chr = gscan_snp.split(":")[0]

        count = 0 # stop after 15 random SNPs found
        for i in range(1, len(sl)): # loop thru all 10K potential random SNPs
            random_snp = sl[i] # chr:pos
            random_chr = random_snp.split(":")[0]

            # random SNP is on same chromosome, not already picked, and in the LIBD genotypes
            if (random_chr == gscan_chr) and \
               (random_snp not in dup_set) and \
               (random_snp in libd_dict):
                dup_set.add(random_snp) # add to our random SNP set so we don't choose it again.
                outline = "{}\t{}\t{}\n".format(gscan_snp, random_snp, libd_dict[random_snp])
                outF.write(outline)
                count += 1
            if count == 15:
                break
        if count < 15: # make sure all 10 random SNPs were found for each of the GSCAN 304 SNPs
            print(gscan_snp)
        line = inF.readline()

````

## LD-prune combined GSCAN and random SNPs
Combine the LD-pruned GSCAN SNPs with the LD-pruned random SNPs. Then perform LD-pruning again to ensure that none of the random SNPs are in LD with the GSCAN SNPs.

Specify that the GSCAN SNPs have preference so that none of them will be removed during the ld-pruning. Then we will have a list of LD-pruned SNPs that contain all of the LD-pruned GSCAN SNPs plus the random SNPs that are not in LD with these GSCAN SNPs nor with each other. We then remove the GSCAN SNPs from this pruned list and we are left with just what we were after - a set of random SNPs that are not in LD with each other nor GSCAN. Then we can take a random sample of 3,050 from this list to get exactly the number we were after.

````r
# combine random LIBD LE SNPs with GSCAN LE SNPs

cat gscan_305/ldprune/libd_matched_with_gscan305_ld_pruned_chrall.prune.in \
    choose_randoms/ldprune/libd_4494_random_matched_ld_pruned_chrall.prune.in \
    > final_random_selection/gscan292_and_3345random_snps.txt # 3,637


# perform LD pruning to make sure random + GSCAN are not in LD
# keep all the GSCAN SNPs
docker run -v $PWD/:/data/ -it rtibiocloud/plink:v2.0_c6004f7 plink2 \
    --memory 2000 \
    --bfile /data/libd_geno/LIBD_NAc_AA \
    --extract /data/final_random_selection/gscan292_and_3345random_snps.txt \
    --indep-pairwise 1500 150 0.2 \
    --indep-preferred /data/gscan_305/ldprune/libd_matched_with_gscan305_ld_pruned_chrall.prune.in \
    --make-bed \
    --out /data/final_random_selection/ldprune/gscan292_and_3345random_snps_ldpruned
    
wc -l final_random_selection/ldprune/gscan292_and_3345random_snps_ldpruned.prune.in # 3,499
    
# remove the GSCAN SNPs from the final merged ldpruned output
grep --invert-match -f gscan_305/ldprune/libd_matched_with_gscan305_ld_pruned_chrall.prune.in \
    final_random_selection/ldprune/gscan292_and_3345random_snps_ldpruned.prune.in > \
    final_random_selection/random_snps3207_ld_pruned.txt 

wc -l final_random_selection/random_snps3207_ld_pruned.txt # 3207

#subset those down to 3050 SNPs (305 unique GSCAN SNPs found in LIBD eQTL results times 10) using a random selection
N=3050
file=final_random_selection/random_snps3207_ld_pruned.txt
shuf -n $N $file > final_random_selection/final_random_snps3050_ld_pruned.txt

wc -l final_random_selection/final_random_snps3050_ld_pruned.txt # 3050

# capture the chromosomes for each of these 3050 SNPs
grep -f final_random_selection/final_random_snps3050_ld_pruned.txt \
    final_random_selection/ldprune/gscan292_and_3345random_snps_ldpruned.bim > \
    final_random_selection/final_random_snps3050_ld_pruned.bim
````

````r
#gscan305 <- read.delim("input_snps_annotated.txt")
gscan305 <- read.delim("SNPsnap_305_gscan_input_v01/input_snps_identifer_mapping.txt")
random3050 <- read.delim("final_random_selection/final_random_snps3050_ld_pruned.bim", header = FALSE)

gscan_chroms <- data.frame(do.call('rbind', strsplit(as.character(gscan305$snpID), ":", fixed=TRUE)), stringsAsFactors=FALSE)
names(gscan_chroms) <- c("chr", "pos")

# add manually the missing SNPs  6:29796685
missing <- data.frame(chr = c(6), pos=c(29796685))
gscan_all <- rbind(gscan_chroms, missing)
gscan_all$chr <- as.integer(gscan_all$chr)
gscan_all$pos <- as.integer(gscan_all$pos)

names(random3050)[1] <- "chr"

# just use dbSNP to find the chr pos
gscan_all$SNPs = "gscan"
random3050$SNPs = "random3050"

combined_data <- rbind(gscan_all[, c("chr", "SNPs")], random3050[, c("chr", "SNPs")])
dim(combined_data)
head(combined_data)
````
