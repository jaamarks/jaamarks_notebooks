## Question 6
Perform PCA on filtered data.

### Read in data

````r
df <- read.table(file = "GSE78280_Non-normalized_data_filtered.txt", header=T, row.names=1)
map_file <- read.table(file = "mapping_file.txt")

head(df)
head(map_file)
````

### Perform PCA

````r
# take transpose since prcomp expects samples in the rows
# center to zero and scale (normalize variables to have unit variance)
pca_data <- prcomp(t(df), center=TRUE, scale. = TRUE)

summary(pca_data)
names(summary(pca_data))
````

````r
pc1 <- summary(pca_data)$importance["Proportion of Variance", "PC1"]
pc1 <- round(pc1 * 100, digits = 2)
pc2 <- summary(pca_data)$importance["Proportion of Variance", "PC2"]
pc2 <- round(pc2 * 100, digits = 2)
#pca_data$sdev^2 / sum(pca_data$sdev^2)*100
# view percent variance explained by each PC
pc1
pc2
````

### Create Plot

````r
### R ###

## create a color coded pca plot
color_map = c()
for (i in 1:length(row.names(pca_data$x))) {
    # remove the X from row names
    mysub = substr(row.names(pca_data$x)[i], 2, nchar(row.names(pca_data$x)[i]))
    index <- which(map_file$V1==mysub)
    if (map_file[index,2] == "MOR") {
        color_map[i] = "red"
    }
    else {
        color_map[i] = "blue"
    }
}


plot(x = pca_data$x[,1], y = pca_data$x[,2], col=color_map,
    main="PCA of Gene Expression: PC1 vs PC2",
    ylab = paste("PC2 (", pc2, "% explained var.)", sep=""),
    xlab = paste("PC1 (", pc1, "% explained var.)", sep=""))

#legend(x=pca_data$x[,1], y=pca_data$x[,2], legend=c(\"MOR\",\"SACH\"), col=c(\"red\", \"blue\"))"
legend("topright", c("MOR","SACH"), col=c("red", "blue"), pch=c(1,1))
````

## Question 7
Discussion.

```
We normalized the data before running the PCA by applying the scale option with the prcomp function. Variables with larger standard deviations will have a higher weight that variables with smaller standard deviations. Certain variables with larger magnitudes may dominate associates between variables in the sample. To ensure that each variable has equal weight in the PCA model, we thus normalized the data prior to the analysis.
```
